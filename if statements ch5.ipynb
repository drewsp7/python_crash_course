```python
#importing files: https://stackoverflow.com/questions/20908018/import-multiple-excel-files-into-python-pandas-and-concatenate-them-into-one-dat 
# import useful libraries 
import sys
import pandas as pd
import os
import re

#get directory and list files
path = os.getcwd()
files = os.listdir(path)
files

# pick out the xlsx files
files_xlsx = [f for f in files if f[-4:] == 'xlsx']
files_xlsx

#initialize empty dataframe df:
df = pd.DataFrame()

#loop over list of xlsx files to put into empty dataframe
for f in files_xlsx:
    data = pd.read_excel(f, 'sales')   #versions that have the data are on tab called 'sales'
    df = df.append(data)

print(df)
```
